```python
import csv
with open("anees_R_2020_08_08_Fish.csv",newline="") as f:
    reader = csv.reader(f)
    data = list(reader)
data[0][0] = "Species"
dataDict = {}
def loadData(data,i):
    lst =list()
    for line in data:
        lst.append(line[i])
    return lst[1:]
for i in range(len(data[0])):
    dataDict[data[0][i]] = loadData(data,i)
for k,v in dataDict.items():
    print(k)
    print(v)
```
